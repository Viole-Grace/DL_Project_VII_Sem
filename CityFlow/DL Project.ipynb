```python
def create_engine_steps():
    eng.reset()
    eng.get_current_time()
    eng.get_lane_vehicle_count()
    print("Lanes : {} \nVehicles {}".format(eng.get_lane_vehicles, eng.get_lane_waiting_vehicle_count))
    eng.get_vehicles(include_waiting=False)
    for i in range(300):
        eng.next_step()
    
    return
```

```python
create_engine_steps()
```

```python
def stop_all_engines():
    ve = eng.get_vehicle_count()
    if ve >1:
        eng.reset()
    
    return
```
